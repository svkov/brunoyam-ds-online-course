```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
```

```python
plt.rcParams['axes.facecolor'] = 'white'
plt.rcParams['figure.facecolor'] = 'white'
```

```python
n = 100
loc = 100
sigma_2 = 10
x = np.arange(n)
y = np.random.normal(loc, sigma_2, size=n)
plt.plot(x, y)
a, b = stats.norm.interval(0.95, loc=loc, scale=sigma_2)
left_bound = a
right_bound = b
plt.fill_between(x, left_bound, right_bound, color='b', alpha=.1)
plt.ylim(50, 150)
plt.ylabel('Кол-во продаж')
plt.xlabel('Номер месяца')
plt.title('95%-ый доверительный интервал')
plt.savefig('../images/conf_int_example.png')
```

```python
uniform_dist = np.random.uniform(0, 1, size=5000)
plt.hist(uniform_dist)
plt.title(f'Равномерное распределение на [0, 1]')
plt.savefig('../images/uniform_clt.png')
```

```python
n = 2
sample = np.random.uniform(0, 1, size=(5000, n))
mean = sample.mean(axis=1)
plt.hist(mean, bins=30)
plt.title(f'Гистограмма средних значений, n={n}')
plt.savefig(f'../images/uniform_clt_n_{n}.png')
```
