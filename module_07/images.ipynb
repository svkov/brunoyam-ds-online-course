# Урок 2_2

```python
def f(x):
    return x * x

def f_prime(x):
    return 2 * x

x = np.linspace(-10, 10)
y = f(x)

plt.plot(x, y)

x_points = np.linspace(-4, 4, 5)
y_points = f(x_points)
plt.scatter(x_points, y_points, color='r')

for x_p in x_points:
    x = np.linspace(x_p - 5, x_p + 5, 1000)
    y_p = f_prime(x_p) * (x - x_p) + f(x_p)
    plt.plot(x, y_p, linewidth=0.5, color='r')
    
plt.legend(['f(x)', 'касательные'])

plt.savefig('../images/parabola_tang.png')
```

```python
def f(x):
    return -x * x

def f_prime(x):
    return -2 * x

x = np.linspace(-10, 10)
y = f(x)

plt.plot(x, y)

x_points = np.linspace(-4, 4, 5)
y_points = f(x_points)
plt.scatter(x_points, y_points, color='r')

for x_p in x_points:
    x = np.linspace(x_p - 5, x_p + 5, 1000)
    y_p = f_prime(x_p) * (x - x_p) + f(x_p)
    plt.plot(x, y_p, linewidth=0.5, color='r')
    
plt.legend(['f(x)', 'касательные'])

plt.savefig('../images/parabola_tang_reversed.png')
```

```python
def f(x):
    return 6 * x ** 2 - x ** 3

def f_prime(x):
    return 12 * x - 3 * x ** 2

x = np.linspace(-15, 15, 1000)
y = f(x)

plt.plot(x, y)
plt.title('Пример точки перегиба')

x_0 = 2
y_0 = f(x_0)

plt.scatter([x_0], [y_0], color='r')

x_t = np.linspace(x_0 - 12, x_0 + 12, 1000)
y_t = f_prime(x_0) * (x_t - x_0) + f(x_0)
plt.plot(x_t, y_t, color='r')

plt.legend(['f(x)', 'касательная'])

plt.savefig('../images/inflection_point.png')
```

# Урок 3

```python
def f(x, y):
    return x * x + y * y - 2 * x + 3 * y - 25

def grad_f(x, y):
    grad_x = 2 * x - 2
    grad_y = 2 * y + 3
    return np.array([grad_x, grad_y])

x0 = np.array([15, 10])
x_old = x0
# Эти параметры можно менять и смотреть на результат
eps = 10e-6
learning_rate = 0.1
max_steps = 100

i = 0
while np.linalg.norm(grad_f(x_old[0], x_old[1])) > eps:
    # Обновляем x
    x_new = x_old - learning_rate * grad_f(x_old[0], x_old[1])
    x_old = x_new
    
    # Увеличиваем счетчик
    i += 1
    # Проверяем, нужно ли выходить из цикла
    if i == max_steps:
        break

print(x_new, f(x_new[0], x_new[1]), i)
```
