```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = [0, 0]
Y = [0, 0]
U = [3, 1]
V = [1, 3]

plt.quiver(X, Y, U, V, units='width')
# plt.xlim(0, 1)
# plt.ylim(0, 1)
```

```python
a = np.array([0, 0])
b = np.array([3, 1])
c = np.array([1, 3])
points = np.array([a, b, c])

plt.arrow(*a, *b, head_width=0.1, head_length=0.2, length_includes_head=True, fc='black')
plt.arrow(*a, *c, head_width=0.1, head_length=0.2, length_includes_head=True, fc='black')

plt.scatter(points[:, 0], points[:, 1], c='black')

plt.annotate('A', (a[0] - 0.25, a[1]), fontsize=16)
plt.annotate('B', (b[0] + 0.1, b[1]), fontsize=16)
plt.annotate('C', (c[0] + 0.1, c[1]), fontsize=16)
plt.grid(True, alpha=0.3)

plt.xlim(-1, 4)
plt.ylim(-1, 4)
plt.savefig('../images/vec_on_plane.png')
```

```python
res = b + c

points = np.array([a, b, res])
plt.arrow(*a, *res, head_width=0.1, head_length=0.2, length_includes_head=True, fc='black')
plt.arrow(*a, *b, head_width=0.1, head_length=0.2, length_includes_head=True, fc='black')
plt.arrow(*b, *c, head_width=0.1, head_length=0.2, length_includes_head=True, fc='black')

plt.scatter(points[:, 0], points[:, 1], c='black')

plt.annotate('A', (a[0] - 0.3, a[1]), fontsize=16)
plt.annotate('B', (b[0] + 0.2, b[1]), fontsize=16)
plt.annotate('AC + AB', (res[0] - 0.5, res[1] + 0.4), fontsize=16)
plt.grid(True, alpha=0.3)
plt.xlim(-0.5, 5)
plt.ylim(-0.5, 5)
plt.savefig('../images/vec_sum.png')
```

```python
points = np.array([a, 2*b, b])
plt.arrow(*a, *b, head_width=0.1, head_length=0.2, length_includes_head=True, fc='black', width=0.02)
plt.arrow(*a, *(2 * b), head_width=0.1, head_length=0.2, length_includes_head=True, fc='black')
plt.scatter(points[:, 0], points[:, 1], c='black')

plt.annotate('A', (a[0] + 0.35, a[1]), fontsize=16)
plt.annotate('B', (b[0] + 0.2, b[1] - 0.1), fontsize=16)
plt.annotate('D', ((2 * b)[0], (2 * b)[1] - 0.2), fontsize=16)
plt.grid(True, alpha=0.3)
plt.savefig('../images/vec_stretch.png')
```
